```python
import pandas as pd
newtrain = pd.read_csv("traindat.csv")
```

```python
newtrain
```

```python
def adjust_value(x):
    if x['loan_default'] > 0:
        return x['no_of_curr_loans']+2
    else:
        return x['no_of_curr_loans']
newtrain['no_of_curr_loans']=newtrain.apply(adjust_value, axis=1)
print(newtrain)
```

```python
newtrain
```

```python
newtrain.groupby(by='loan_default').agg('mean')[['loan_amount','age','education','no_of_loans','asset_cost','no_of_curr_loans','last_delinq_none']]
```
